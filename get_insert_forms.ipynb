```python
forms_path = "/Volumes/workspace/anima/anima_volume/raw/dynamic_ab_research/forms.csv"

forms_df = (
    spark.read
         .option("header", True)
         .option("inferSchema", True)
         .csv(forms_path)
)

display(forms_df.limit(5))
```

```python
forms_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("workspace.anima.forms")
```
